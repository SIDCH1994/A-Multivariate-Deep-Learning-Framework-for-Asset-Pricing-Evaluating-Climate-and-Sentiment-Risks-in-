# Daily Temp & Precipitation via meteostat

```python
import pandas as pd
import os
from datetime import date
import meteostat as ms

# 1. Directory Setup
base_dir = r"D:\MS_Data_Science_Thesis\Data_Extraction"
raw_folder = os.path.join(base_dir, "Raw_Data_Folder")

def extract_lubbock_climate():
    print("--- Starting Texas High Plains Climate Extraction (Meteostat v2) ---")
    
    # 2. Define Timeframe (v2 uses 'date' instead of 'datetime')
    start = date(2010, 1, 1)
    end = date(2025, 12, 31) 
    
    # 3. Define Location (Lubbock, TX coordinates)
    lubbock_point = ms.Point(33.6666, -101.8228, 999)
    
    # 4. Fetch Daily Data
    print("Fetching daily meteorological data for Lubbock, TX...")
    
    # Get the 4 nearest weather stations to the Lubbock coordinates
    stations = ms.stations.nearby(lubbock_point, limit=4)
    
    # Fetch the daily data and interpolate it for our exact geographic point
    ts = ms.daily(stations, start, end)
    df = ms.interpolate(ts, lubbock_point).fetch()
    
    if df.empty:
        print("Error: No weather data found.")
        return
        
    # 5. Clean and Filter
    # In v2, parameters are accessed via Enums. We want Max Temp (TMAX) and Precipitation (PRCP)
    target_columns = [ms.Parameter.TMAX, ms.Parameter.PRCP]
    df = df[target_columns].copy()
    
    # Reset index to turn the Date index into a standard column
    df.reset_index(inplace=True)
    
    # Rename columns to match your thesis notation and avoid Enum objects
    df.columns = ['Date', 'Max_Temp_C', 'Precipitation_mm']
    
    # 6. Export to CSV
    output_path = os.path.join(raw_folder, "lubbock_climate_raw.csv")
    df.to_csv(output_path, index=False)
    
    print(f"Success! Climate data exported to: {output_path}")
    print(f"Data Shape: {df.shape}")

if __name__ == "__main__":
    extract_lubbock_climate()
```
