# U.S. Dollar Index (DXY)

### How DXY Affects Cotton Prices
Because the ICE Cotton No. 2 contract is globally benchmarked and priced exclusively in U.S. dollars, it has a rigid, inverse mathematical relationship with the DXY.

The United States is a massive exporter of cotton, shipping the majority of its crop to foreign textile hubs. When the U.S. dollar strengthens (the DXY goes up), U.S. cotton becomes mathematically more expensive for foreign buyers to purchase with their local currencies, even if the actual physical supply and demand haven't changed. This artificial price inflation can cause foreign mills to source cheaper cotton from competitors like Brazil or India, which tanks the U.S. futures price. Conversely, a weak dollar makes U.S. cotton cheaper globally, driving up export demand and pushing futures higher.

Including the DXY allows your neural network to distinguish between a price drop driven by an actual agricultural surplus and one driven purely by macroeconomic currency fluctuations.

```python
import yfinance as yf
import pandas as pd
import os

# 1. Directory Setup
base_dir = r"D:\MS_Data_Science_Thesis\Data_Extraction"
raw_folder = os.path.join(base_dir, "Raw_Data_Folder")

if not os.path.exists(raw_folder):
    os.makedirs(raw_folder)

def extract_dxy_data():
    print("--- Starting U.S. Dollar Index (DXY) Extraction ---")
    
    # 2. Define Ticker and Timeframe
    ticker = "DX-Y.NYB"
    start_date = "2010-01-01"
    end_date = "2025-12-31"
    
    # 3. Download Data
    print(f"Downloading {ticker} from {start_date} to {end_date}...")
    df = yf.download(ticker, start=start_date, end=end_date, auto_adjust=False)
    
    if df.empty:
        print(f"Error: No data found for {ticker}.")
        return

    # 4. Handle MultiIndex columns
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    print("Columns found:", df.columns.tolist())

    # 5. Basic Cleaning
    df = df.dropna(subset=['Close'])

    # 6. Export to CSV
    output_path = os.path.join(raw_folder, "dxy_raw.csv")
    df.to_csv(output_path, index=True)
    
    print(f"Success! DXY data exported to: {output_path}")
    print(f"Data Shape: {df.shape}")

if __name__ == "__main__":
    extract_dxy_data()
```
